# Datasets
Customer Feedback Data

Description: Contains customer reviews and feedback from surveys or social media.
Customer Metadata Data

Description: Contains demographic and interaction details about customers.
Product/Service Information Data

Description: Contains details about the products or services that are being reviewed.


# Dataset Column Descriptions
1. Customer Feedback Data

Feedback_ID:     Unique identifier for each feedback entry.
Customer_ID:     Unique identifier for the customer providing feedback.
Product_ID:      Unique identifier for the product/service being reviewed.
Feedback_Type:   Type of feedback (e.g., Survey, Social Media, Email).
Feedback_Date:   Date when the feedback was given.
Rating:          Rating given by the customer (1 to 5 scale).
Comment:         Text of the feedback comment.
Sentiment_Label:  Sentiment of the feedback (Positive, Neutral, Negative).

2. Customer Metadata Data

Customer_ID:      Unique identifier for each customer.
Name:             Full name of the customer.
Age:              Age of the customer.
Gender:           Gender of the customer.
Location:         Location where the customer resides.
Date_of_Birth:     Birth date of the customer.
Registration_Date:  Date when the customer registered.

3. Product/Service Information Data

Product_ID:       Unique identifier for each product/service.
Product_Name:     Name of the product/service.
Category:        Category of the product/service.
Price:           Price of the product/service.
Manufacturer:     Manufacturer of the product/service.
Launch_Date:      Date when the product/service was launched.
Description:       Description of the product/service.

# 4. Data Merging


# Merge the datasets using common columns (Customer_ID and Product_ID)

```python
merge_data=pd.merge(customer_feedback_data,customer_metadata__data,on='Customer_ID')
merge_data=pd.merge(merge_data,product_information_data,on='Product_ID')
```

```python
merge_data.columns
```

# 5. Sentiment Distribution Analysis

# Perform a basic sentiment analysis by plotting sentiment distribution

```python
sns.countplot(x='Sentiment_Label',data=merge_data,palette=['orange','green','red'])
plt.title("Sentiment Distribution",fontsize=20)
plt.show()
```

# Analyze the sentiment distribution across different product categories

```python
sns.countplot(x='Category',hue='Sentiment_Label',data=merge_data)
#plt.xticks(rotation=90)
plt.title("sentiment by product category",fontsize=20)
```

# 6.Feedback Analysis

```python

sns.countplot(x='Feedback_Type',data=merge_data,palette=['orange','green','red'])
plt.title("feedback analysis",fontsize=20)
plt.show()
```

#  7.Customer Demographics Impact


# Analyze sentiment distribution based on demographics like age and gender:


```python
# Sentiment vs. Age
sns.boxplot(data=merge_data, x='Sentiment_Label', y='Age',palette=['orange','red','green'])
plt.title('Sentiment by Age Group',fontsize=20)
plt.show()
```

# Examine sentiment distribution by Gender

```python
sns.countplot(x='Gender',hue='Sentiment_Label',data=merge_data)
plt.title("sentiment distribution for gender",fontsize=20)
```

# 8. Product/Service Analysis

# 9.Time Series Analysis

```python


# Convert 'Feedback_Date' to datetime format
merge_data['Feedback_Date'] = pd.to_datetime(merge_data['Feedback_Date'])

# Optional: Filter the data to focus on specific product or service (if needed)
# df = df[df['Product_Name'] == 'Product X']  # Example for filtering specific product

# Convert Sentiment_Label into numeric values for easier analysis
# Example: Positive = 1, Neutral = 0, Negative = -1
merge_data['Sentiment_Score'] = merge_data['Sentiment_Label'].map({'Positive': 1, 'Neutral': 0, 'Negative': -1})

# Group the data by Feedback_Date and calculate the average sentiment per day
daily_sentiment = merge_data.groupby('Feedback_Date')['Sentiment_Score'].mean()

# Plot the time series of average sentiment scores over time
plt.figure(figsize=(10,6))
plt.plot(daily_sentiment, color='green', marker='o', linestyle='-', label='Average Sentiment Score')
plt.title('Sentiment Trend Over Time',fontsize=20)
plt.xlabel('Date')
plt.ylabel('Average Sentiment Score')
plt.grid(True)
plt.legend()
plt.show()
```

# 9.Create visualizations such as bar charts, and sentiment heatmaps to represent findings.

 # Bar Chart for Sentiment Distribution

```python
# Count the occurrences of each sentiment label
sentiment_count = merge_data['Sentiment_Label'].value_counts()
plt.figure(figsize=(6,4))
sentiment_count.plot(kind='bar', color=['green', 'gray', 'red']) 
plt.xlabel("Sentiment", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Sentiment Distribution", fontsize=20)  
plt.xticks(rotation=0)
plt.show()

```

```python


# Ensure 'Feedback_Date' is in datetime format
#merge_data['Feedback_Date'] = pd.to_datetime(merge_data['Feedback_Date'])

# Create a 'Month' column
merge_data['Month'] = merge_data['Feedback_Date'].dt.to_period('M')

# Group by Month and Sentiment_Label, counting each sentiment
sentiment_by_month = merge_data.groupby(['Month', 'Sentiment_Label']).size().unstack(fill_value=0)

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(sentiment_by_month, cmap='coolwarm', annot=True, fmt="d")
plt.title('Sentiment Heatmap by Month')
plt.show()
```
